```python
#map function
nums = [1,2,3,4,5]
def sq(n):
    return n*n
square = map(sq,nums)
print(list(square))
```

```python
#zip function
name = ["maohan", "das", "Ram"]
ID = [4,8,5]
mapped= zip(name, ID)
print(list(mapped))
```

```python
#loops 
```

```python
Fruits = ["apple", "banana", "cheery", "pineapple"]
for x in Fruits:
    print(x)
```

```python
for x in Fruits:
    print(x)
    if x == "cherry":
     break
```

```python
for x in Fruits:
    if x == "pineapple":
        continue
    print(x)
```

```python
Range = [1,2,3,4,5,6]
for i in Range:
    print(i)
```

```python
Range = [1,2,3,4,5,6]
for i in Range:
    if i ==  4:
        continue
    print(i)
```

```python
#while loop
```

```python
num = 1
while num < 6:
    print(num)
    num += 1
```

```python
num = 1
while num < 10:
    print(num)
    if num == 5:
        break
    num += 1
```

```python
num = 0
while num < 7:
    num += 1
    if num == 5:
        continue
    print(num)
```
